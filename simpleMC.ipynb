```python
import numpy as np
import pylab as plt
plt.rcParams['figure.figsize'] = [7, 7]
from tqdm.notebook import tqdm
```

```python
N=2000

samples = np.random.uniform(-1,1,N*2).reshape((2,N))
```

```python
plt.scatter(samples[0],samples[1])
```

```python
dist = np.sum(samples**2,axis=0)
```

```python
goodones = samples.T[dist<1].T
```

```python
plt.scatter(goodones[0],goodones[1])
```

```python
# 4 is the area of the square
mypi = 4* goodones.shape[1]/samples.shape[1]
```

```python
mypi
```

```python
def findpi(N):
    samples = np.random.uniform(-1,1,N*2).reshape((2,N))
    dist = np.sum(samples**2,axis=0)
    goodones = samples.T[dist<1].T
    return 4* goodones.shape[1]/samples.shape[1]
```

```python
Nvals=np.logspace(8,1,100).astype(int)
```

```python
pivals = np.array([findpi(N) for N in tqdm(Nvals)])
```

```python
plt.plot(Nvals,pivals)
plt.axhline(np.pi,c='red',ls='dashed')
#plt.ylim(np.pi,np.pi)
plt.semilogx()
```
